```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Call_Logs.csv')
df.head()
```

```python
df.Logs[0]
```

```python
#Define a function to extract date, time and conversation
def extract_info(df):
    #Split the info by \n
    lines = df['Logs'].split('\n')
    #extract line time
    date = lines[0].split(': ')[1]
    time = lines[1].split(': ')[1]
    #Extract converstaion
    conv = "\n".join([line for line in lines[3:] if line != ""])
    return date, time, conv
```

```python
#Apply extract_info function to each row in the data frame to extract the key inforamtions
df[['date','time','conv']] = df.apply(extract_info, axis=1, result_type='expand')
df.head()
```

```python
df.conv[0]
```

```python
#Remove Logs and unnamed columns from the data frame
df.drop(["Unnamed: 0","Logs"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-base")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-base")

#empty list to store the summaries
output_conv = []

for conv in df['conv']:
    #Prepare prompt for summarization
    prompt = f"""
    {conv}
    What were the main points in that conversation
    """
    #tokenize the prompt and generate the summary
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids)
    summary = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    # append the summary to a the list
    output_conv.append(summary)
```

```python
df["Summary"] = output_conv
df.head()
```

```python
df['Summary']
```

```python
#load zero-shot classification pipeline from the transformers library t find the cancellation and other types of conversation
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/multilingual-MiniLMv2-L6-mnli-xnli")
#Define labels
labels = ['cancellation','other']

#declare empty list
classification = []

for conv in df['conv']:
    result = classifier(conv, labels)
    classification.append(result['labels'][0])

df["Cancellation"] = classification

df["Cancellation"]=[True if cls == 'cancellation' else False for cls in classification]
df.head()
```

```python
df.head()
```

```python
#Function to identify reason for cancellation
def cancellation_reason(df):
    """Check if the conversation is flagged for cancellation"""
    if df['Cancellation']==False:
        return None
    else:
        # Create a prompt
        prompt = f"""
        {df['conv']} 
        What are the issues that led the client to cancel the subscription
        """
        # Convert the prompt to tokens that send to the model
        input_ids = tokenizer(prompt, return_tensors="pt").input_ids
        output_ids = model.generate(input_ids, max_new_tokens=50, min_length=20)
        return tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

```python
df["Cancellation Reasons"] = df.apply(cancellation_reason, axis=1)
df
```

```python
#Print all non-None reasons
for reason in df["Cancellation Reasons"]:
    if reason != None:
        print(reason)
```

```python
from wordcloud import WordCloud, STOPWORDS

def generate_wordcloud(text_series, title):
    #Convert text data into a single string
    text = " ".join(review for review in text_series)
    #Generate word Cloud image
    word_cloud = WordCloud(width=800, height=400, background_color="white").generate(text)
    #setup the plot parameter
    plt.figure(figsize=(10,5))
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python
#Generate the word cloud for conversation
generate_wordcloud(df['conv'],"Word Cloud for Conversation")
```

```python
#Generate the world cloud for summary
generate_wordcloud(df['Summary'], "Word Cloud for Summary")
```
